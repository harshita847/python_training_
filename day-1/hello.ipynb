```python
#write a program which will take a no. input from user and if the no.is even it will calcualte even factorial and if odd then odd factorial

def even_fact(n):
    ans=1
    for i in range(2,n+1,2):
        ans*=i
    return ans

def odd_fact(n):
    ans=1
    for i in range(1,n+2,2):
        ans*=i
    return ans

n=int(input("enter no:"))
if(n%2==0):
    print(even_fact(n))
else:
    print(odd_fact(n))
```
