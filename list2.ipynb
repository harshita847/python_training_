```python
#wap which will create a list with all the element divisible by 3 in between 4 to 25

l1=[]
for i in range(3,25,3):
    l1.append(i)
print(l1)

# list comphrension
l2=[i for i in range(4,25) if i%3==0]
print(l2)
```

```python
l1=[i for i in range(4,25) if i%3==0]

print(l1)

l2=[i*i for i in l1 if i%2==0]
print(l2)

l3=[i*i for i in l1 if i%2!=0]
print(l3)
```
